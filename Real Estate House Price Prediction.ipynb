```python
# This function gives descriptive information about columns such as column name,type,null values existence and so on
def analyzeColumns(dataframe):
    print('Analyzing dataframe... \n')
    
    # Checking type of the dataframe
    if type(dataframe) is pd.core.frame.DataFrame:
        print('Dataframe type is correct\n')
    else:
        print('Wrong dataframe type\n')
        return False
        
    columns = list(dataframe.columns)
    for (index,column) in enumerate(columns):
        print('------ Column ' + str(index+1)  + ' ------')
        print('Name: ' + column)
        
        coltype = str(dataframe[column].dtype)
        
        # Checking type of the column
        print('Type: ' + coltype)
        print('Index: ' + str(index))
        # Checking if the dataframe contains any null values 
        print('Null values exist: ' + ('Yes' if dataframe.isna().any().any() else 'No'))
        
        # If column contains numeric values compute the statistical informations
        if(coltype == 'int64' or coltype == 'float64'):
            print('Value Range: (' + str(dataframe[column].unique().min()) + ',' + str(dataframe[column].unique().max()) + ')')
            print('Distance MinMax: ' +  str(dataframe[column].unique().max()- dataframe[column].unique().min()))
            print('Mean: ' + str(dataframe[column].mean()))
            print('Std Dev: ' + str(dataframe[column].std()))
        # If column is date display with necessary formatting
        elif(coltype == 'datetime64[ns]'):
            ts1 = pd.to_datetime(str(dataframe[column].unique().min()))
            ts2 = pd.to_datetime(str(dataframe[column].unique().max()))
            minday = ts1.strftime('%d/%m/%Y')
            maxday = ts2.strftime('%d/%m/%Y')
            print('Value Range(d/m/y): (' + minday + ',' + maxday + ')')
        
         
        print('-----------------------\n')
    print('Finished analyzing dataframe... \n')    
    return True
```

```python
def get_prediction(message):
    
    print('inside get prediction method') 
    #print(message['intent'])
    
    testdf = pd.DataFrame()
    
    for key,val in message['fields'].items():
        #print(key + ': ' + str(val) + '\n')
        testdf[key] = pd.Series(data = float(val))
    
    #print(testdf.iloc[0])
    testdf  = pd.concat([testdf, df_for_pred ],ignore_index=True)
    testdf = testdf.drop(columns=['price'])
    #print(testdf.head(10))
    
    columns = list(testdf.columns)
    
    for (index,column) in enumerate(columns):
        coltype = str(testdf[column].dtype)
        #print('\ncoltype is ' + coltype + '\n')
        x2 = testdf[[column]].values #returns a numpy array
        min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1))
        x_scaled2 = min_max_scaler.fit_transform(x2)
        data5 = pd.DataFrame(x_scaled2)

        testdf = testdf.drop([column,column], axis=1)

        testdf.insert(0,column,data5)
    
    #print(testdf.head(10))
    price_val = regr.predict(testdf.iloc[0].to_frame().transpose()).item(0)
  
    #print('\nEstimated price is: ' + str(price_val) + '\n')
    
    
    
    return price_val
    
```

```python
import websocket
try:
    import thread
except ImportError:
    import _thread as thread
import time



def on_message(ws, message):
    #print(message)
    message = json.loads(message)
    
    if('intent' in message and message['intent'] == 'Get_Prediction'):
        print('Inside Watson Studio... get prediction detected... ')
        predicted_val = get_prediction(message)
        print(predicted_val)
        prediction = {"prediction": predicted_val}
        print('\ndict: ' + ''.join(prediction) + ' ' + str(prediction['prediction']) + '\n')
        print('\nbefore ws send \n')
        ws.send(json.dumps(prediction))

def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")

def on_open(ws):
    def run(*args):
        '''for i in range(3):
            time.sleep(1)
            ws.send("Hello %d" % i)
        time.sleep(1)
        ws.close()
        '''
        print("on open method started...")
    thread.start_new_thread(run, ())


def start_websocket_listener():
    websocket.enableTrace(True)
    ws = websocket.WebSocketApp("ws://<your ip address>:<your port number>/<your websocket path>",
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close)
    ws.on_open = on_open
    ws.run_forever()
```
